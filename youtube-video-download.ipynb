```python
import yt_dlp

link = input("Video link: ")

ydl_opts = {
    'format': 'best',
    'outtmpl': '%(title)s.%(ext)s',
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([link])

print("Video downloaded successfully")
```

```python
import yt_dlp

link = input("Video link: ")

ydl_opts = {
    'format': 'bestaudio/best',  # En yüksek ses kalitesini seç
    'postprocessors': [{  # Ses dosyasını MP3 formatına dönüştür
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',  # MP3 kalitesi (kbps)
    }],
    'outtmpl': '%(title)s.%(ext)s',  # Dosya adını video başlığı ile adlandır
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([link])

print("Audio downloaded successfully")
```
